# Face Recognition with OpenCV's built-in Haar Cascades ðŸ“¸

```python
import cv2
import matplotlib.pyplot as plt
```

```python
cv2.__version__
```

```python
# Load the cascade
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

Other classifier XML files can be downloaded from OpenCV [repo](https://github.com/opencv/opencv/tree/master/data/haarcascades)

```python
# Read the sample image
img = cv2.imread('yann.jpeg') 
type(img) # => numpy.ndarray
```

```python
# Sometimes it helps to convert image into grayscale (many OpenCV operations are performed on grayscale by default):
# gray_img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
# Detect faces
faces = face_cascade.detectMultiScale(img, 1.1, 4)
```

```python
# Draw rectangle around the faces
for (x, y, w, h) in faces:
    cv2.rectangle(img, (x, y), (x+w, y+h), (0, 0, 255), 2)
```

```python
plt.imshow(img)
```

As you can see, your RGB colours got mixed up with BGR ðŸ˜® A quick way to fix that is to convert the image from RGB to BGR, to simply switch which value corresponds to which colour.

```python
# Display the end result
rgb_img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
type(rgb_img) # => numpy.ndarray
plt.imshow(rgb_img)
```
